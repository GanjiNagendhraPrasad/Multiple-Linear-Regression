```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/data.csv')
df.head()
```

```python
df.shape
```

```python
df=df.drop(['date','street','statezip'],axis=1)
```

```python
df.columns
```

```python
len(df['city'].unique())
```

```python
f = {}
s = 0

for i in df['city']:
  if i not in f:
    f[i] = s
    s = s + 1
  else:
    pass
df['city'] = df['city'].map(f).astype(int)
```

```python
df['city'].unique()
```

```python
df.head()
```

```python
df['country'].unique()
```

```python
len(df['country'].unique())
```

```python
df['country']=df['country'].map({'USA':1}).astype(int)
```

```python
df.head()
```

```python
df['country'].unique()
```

```python
df.columns
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
m=LinearRegression()
m.fit(x_train,y_train)
```

```python
m.coef_
```

```python
m.intercept_
```

```python
ans_train=m.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(f'mse {mean_squared_error(y_train,ans_train)}')
print(f'train acc {r2_score(y_train,ans_train)}')
```

```python
ans_test=m.predict(x_test)
print(f'mse {mean_squared_error(y_test,ans_test)}')
print(f'test acc {r2_score(y_test,ans_test)}')
```

```python
class MLR:
  def __init__(self,d):
    self.df=pd.read_csv(d)
    self.df=self.df.drop(['date','street','statezip'],axis=1)
    f = {}
    s = 0
    for i in self.df['city']:
      if i not in f:
        f[i] = s
        s = s + 1
      else:
        pass
    self.df['city'] = self.df['city'].map(f).astype(int)
    self.df['country']=self.df['country'].map({'USA':1}).astype(int)
    self.x=self.df.iloc[:,1:]
    self.y=self.df.iloc[:,0]
    self.x_train,self.x_test,self.y_train,self.y_test=train_test_split(self.x,self.y,test_size=0.2,random_state=42)

  def train(self):
    self.m=LinearRegression()
    self.m.fit(self.x_train,self.y_train)
    self.ans_train=self.m.predict(self.x_train)
    print(f'mse {mean_squared_error(self.y_train,self.ans_train)}')
    print(f'train acc {r2_score(self.y_train,self.ans_train)}')
  def test(self):
    self.ans_test=self.m.predict(self.x_test)
    print(f'mse {mean_squared_error(self.y_test,self.ans_test)}')
    print(f'test acc {r2_score(self.y_test,self.ans_test)}')
if __name__=="__main__":
  obj=MLR('/content/data.csv')
  obj.train()
  obj.test()
```
